This means, that out of a list of pairs of words for a subclass, there should be a value assigned to each pair corresponding to how well it fits the subclass. As with most machine learning datasets, there is a `training` and `testing` dataset. The `training` dataset contained 10 of the sub-classes, while the `testing` dataset contained the remaining 69. I found this particularly interesting because it shows that well-scoring models are identifying *new* relationships between pairs of words instead of identifying those that it learned during the training process. Now that we understand the data provided, we can get into the scoring process. The type of question that was asked in `Question 2` of the second part of the survey is a [MaxDiff](https://en.wikipedia.org/wiki/MaxDiff) question, which is supposedly robust to biases that can occur when making an explicit ranking of four pairs of words. As such, the prototypicality score was calculated as: \[Percentage of the time a pair is most illustrative\] - \[Percentage of the time a pair is least illustrative\].

An example of the prototypicality scores are available [here](SemEval-2012-Gold-Ratings.zip). Below are the results for the training sub-category of 2a:
    
<blockquote>
 50.0 "weapon:spear"<br>
 38.8 "bird:robin"<br>
 38.8 "tree:oak"<br>
...<br>
-46.9 "oak:tree"<br>
-49.0 "sweater:knit"<br>
-55.1 "wheat:bread"<br>
</blockquote>

Note that only the training sub-categories had these results available, they are not available for the testing sub-categories.

Now suppose you have trained a natural language model (such as the one we've covered in class) to give a rating for each pair of words provided (such as those from part one of the survey) as to the "degree of relational similarity" "to the relation expressed in that category" (see [paper](https://aclanthology.org/S12-1047.pdf)). The provided relation could have been something like {car:engine, face:nose, novel:epilogue} (the paradigmatic examples) and/or "a Y is a part of X" (the relationship). One metric that was used to score how well a model performs is [Spearman's rank correlation coefficient](https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient) in conjunction with the model's MaxDiff score for pairs. The Wikipedia page for Spearman's rank correlation coefficient has a really good example demonstrating how it's calculated. All that you really need to know, however, is that it returns values from -1 to 1, with 1 meaning that the two lists of prototypicality scores were identical in *order* (not necessarily value), and -1 means that they were completely opposite. Another metric that is mentioned in the [paper](https://aclanthology.org/S12-1047.pdf) is measure the accuracy of a model's scores for MaxDiff pairs. This is done by measure the percentage of the time the model chose the right answers, where the "right answer" is the most commonly selected word pair for most/least illustrative by those who completed the original survey.

There are two baseline measurements provided in the dataset's [paper](https://aclanthology.org/S12-1047.pdf): random and PMI.<br>
<br>
Random should expect the Spearman correlation to be zero and the MaxDiff accuracy score to be 31%. Note that this is *not* 25% since there is a possibility for the original survey results to have a single pair of words to be chosen as both the *most* and *least* illustrative. Since it is possible for two pairs of words to tie for either *most* or *least* illustrative, the [paper](https://aclanthology.org/S12-1047.pdf) states that a truly random method would expect to be right 31% of the time based on the results from the survey (at least, that's how I interpreted this result, if I misunderstood please let me know!)<br>
<br>
PMI gives a prototypicality rating for each word pair from the following formula: $\log_2\left(\frac{p(x,y)}{p(x)p(y)}\right)$ where $p(\cdot)$ is the probability of a word appearing from a corpus of "approximately 50 billion tokens" and $p(\cdot,\cdot)$ is the probability of a two words appearing within a ten-word window from a corpus of "approximately 50 billion tokens". The resulting rating is then used to obtain the *most* or *least* illustrative pairs of words for each sub-category by ordering the pairs of words based on their PMI. Then, the MaxDiff accuracy metric can then be computed based on the resulting *most* or *least* illustrative pairs.

Next, I define some helper functions that I will use in my EDA. I will first explore the first phase of the survey by analyzing the pairs of words collected.

```python
# Returns the number of lines in the sub-class file, the unique words, and the unique pairs of words
def get_lines_words(dataset, ID):
    with open(f'/home/alexey/Datasets/SemEval2012/{dataset}/Phase1Answers/Phase1Answers-{ID}.txt') as f:
        lines = 0 # Count number of lines
        words = set() # Keep track of unique words
        pairs = set() # Keepo track of unique pairs
        for line in f.readlines():
            lines += 1 # Increment line count
            line = line.strip() # Remove white space and newlines
            line = line.replace("\"","") # Remove quotation marks
            pairs = pairs.union({line})
            word1, word2 = line.split(":")
            words = words.union({word1}, {word2})
        return lines, list(words), list(pairs)
    raise Exception("get_lines_words: File cound not be opened")

print(get_lines_words('Training','1a'))
```

```python
# Returns two arrays, showing the number of times a word pair was chosen as most and least illustrative
#   respectively

def get_illustrative_counts(dataset, ID):
    with open(f'/home/alexey/Datasets/SemEval2012/{dataset}/Phase2Answers/Phase2Answers-{ID}.txt') as f:
        _, _, pairs = get_lines_words(dataset,ID) # Get number of word pairs
        most = np.zeros(len(pairs)) # Keep track of most illustrative counts
        least = np.zeros(len(pairs)) # Keep track of least illustrative counts
        for line in f.readlines():
            if line[0] == '#':
                continue
            line = line.replace("\"","")# Remove quotation marks
            line = line.split()[0:4]
            least[pairs.index(line[0])] += 1
            most[pairs.index(line[3])] += 1
        return most, least
    raise Exception("get_word_pairs: File cound not be opened")

print(get_illustrative_counts('Training','1a'))
```

However, this would likely be standardized since one can imagine that a semantic relationship would be approximately the same difficulty across different models.

Next, I explore the second phase of the survey, where pairs of words were ranked according to how illustrative they were.

```python
for dataset_str, dataset_IDs, fig_sz in [("Training", training_IDs, [10,6]), ("Testing", testing_IDs, [16,10])]:
    most_percent = np.zeros(len(dataset_IDs))
    least_percent = np.zeros(len(dataset_IDs))
    for i in range(len(dataset_IDs)):
        res = get_illustrative_counts(dataset_str,dataset_IDs[i])
        most_percent[i] = np.max(res[0])/np.sum(res[0])*100
        least_percent[i] = np.max(res[1])/np.sum(res[1])*100

    # Plot our bar chart
    plt.figure(figsize=[16,10])
    plt.bar(range(len(dataset_IDs)), most_percent, 0.9, color='darkgoldenrod')
    plt.xticks(range(len(dataset_IDs)), dataset_IDs, rotation='vertical')
    plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
    plt.ylabel("Percentage of all answers that were the overall most illustrative")
    plt.xlabel("Sub-class ID")
    plt.title(f"Percentage of answers that were the overall most illustrative for {dataset_str.lower()} dataset")
    plt.show()
    plt.close()

    minimum, median, maximum = np.percentile(most_percent, [0,50,100])
    print("Most %:")
    print(f"Minimum: {minimum}")
    print(f"Median: {median}")
    print(f"Maximum: {maximum}")
    print(f"Variance: {np.var(most_percent)}")
    
    # Plot our bar chart
    plt.figure(figsize=[16,10])
    plt.bar(range(len(dataset_IDs)), least_percent, 0.9, color='darkgoldenrod')
    plt.xticks(range(len(dataset_IDs)), dataset_IDs, rotation='vertical')
    plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
    plt.ylabel("Percentage of all answers that were the overall least illustrative")
    plt.xlabel("Sub-class ID")
    plt.title(f"Percentage of answers that were the overall least illustrative for {dataset_str.lower()} dataset")
    plt.show()
    plt.close()
    
    minimum, median, maximum = np.percentile(least_percent, [0,50,100])
    print("Least %:")
    print(f"Minimum: {minimum}")
    print(f"Median: {median}")
    print(f"Maximum: {maximum}")
    print(f"Variance: {np.var(least_percent)}")
```

For the training dataset, it's shown that the most and least illustrative pair was chosen approximately the same percentage of the time across all sub-classes. However, the percentage is still small, right around 5-6%. This small percentage is similar to the testing dataset, however it has a lot more variance. The range is about 4-8%. This means that some sub-classes yielded more agreement between the survey takers than others. This set of results could explain why the RNN-1600 model presented in class with a MaxDiff of about 0.42 was state of the art at the time: it's much harder in a language dataset to come to an agreement of what an illustrative example of a concept is than in an image dataset. It's easier for people to say "yep, that's a stop sign" than it is for people to say "pet:dog is the best example of Y is a kind/type/instance of X from the pairs of words provided".
